## Modification des scénarios 

Il faut modifier le fichier `./scenarios/beergame_ql_ob.json`

Dedans il y a un scenario `root > smBeergameQlOB > scenarios > train agents`

On peut modifiier les propriétés. 

```python
import json
from IPython.display import JSON

TRAINING_FILE_PATH = './scenarios/beergame_ql_ob.json'
f = open(TRAINING_FILE_PATH)
data = json.load(f)

JSON(data['smBeergameQlOB']['scenarios']['train_agents'])
```

```python
def update_training_parameter_scenario(
    TRAINING_FILE_PATH,
    starttime=1, stoptime=24, dt=1,
    backorder_item_cost=1,
    inventory_item_cost=0.5,
    minimum_inventory_cost=200,
    target_inventory=400,
    weekly_order=400,
):
    with open(TRAINING_FILE_PATH, 'r') as f:
        data = json.load(f)
    
    data['smBeergameQlOB']['scenarios']['train_agents']['runspecs']['starttime'] = starttime
    data['smBeergameQlOB']['scenarios']['train_agents']['runspecs']['stoptime'] = stoptime
    data['smBeergameQlOB']['scenarios']['train_agents']['runspecs']['dt'] = dt
    data['smBeergameQlOB']['scenarios']['train_agents']['properties']['backorder_item_cost']['value'] = backorder_item_cost
    data['smBeergameQlOB']['scenarios']['train_agents']['properties']['inventory_item_cost']['value'] = inventory_item_cost
    data['smBeergameQlOB']['scenarios']['train_agents']['properties']['minimum_inventory_cost']['value'] = minimum_inventory_cost
    data['smBeergameQlOB']['scenarios']['train_agents']['properties']['target_inventory']['value'] = target_inventory
    data['smBeergameQlOB']['scenarios']['train_agents']['agents'][0]['properties']['weekly_order']['value'] = weekly_order
    
    with open(TRAINING_FILE_PATH, "w") as f:
        json.dump(data, f)
```

### 1. Update parameters

```python
import json
from IPython.display import JSON
TRAINING_FILE_PATH = './scenarios/beergame_ql_ob.json'

update_training_parameter_scenario(
    TRAINING_FILE_PATH,
    inventory_item_cost=1,
    backorder_item_cost=1
)

f = open(TRAINING_FILE_PATH, 'r')
data = json.load(f)
JSON(data['smBeergameQlOB']['scenarios']['train_agents'])
```

### 2. Train

```python
bptk.train_scenarios(
    episodes=100,
    scenario_managers=["smBeergameQlOB"],
    scenarios=["train_agents"],
    agents=["controlling"],
    agent_states=["active"],
    agent_properties=["supply_chain_reward"],
    agent_property_types=["total"],
    return_df=False,
    progress_bar=True
)
```
